```
cursor.execute('''SHOW tables''')
tables = cursor.fetchall()
tables = [str(i[0].decode('utf-8')) for i in tables]
print(tables)
  
```

```
#n√∫mero de registros por tabal
for i in tables:
    table = i
    print(table)
    cursor.execute(f'''SELECT count(*)
                    FROM {table}''')
    rows = cursor.fetchone()
    print(rows)
```

```
#numero de columnas por cada tabla
for i in tables:
    table = i
    print(table)
    cursor.execute(f'''SELECT count(*)
                    FROM INFORMATION_SCHEMA.COLUMNS
                    WHERE TABLE_NAME = '{table}'
                    ORDER BY ORDINAL_POSITION''')
    number_columns = cursor.fetchone()
    print(number_columns)
```
